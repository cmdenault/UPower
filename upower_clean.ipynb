```python
# function that given a data frame like pre_pre_post, will compute the average question_score for all question_id values

def calculate_average_question_score(df):
  """
  This function calculates the average question score for each question_id in a DataFrame.

  Args:
      df (pd.DataFrame): The DataFrame containing participant data.

  Returns:
      dict: A dictionary where keys are question_id values and values are the corresponding average question scores.
  """

  average_scores = {}
  for question_id in df['question_id'].unique():
    question_data = df[df['question_id'] == question_id]
    average_score = question_data['question_score'].mean()
    average_scores[question_id] = average_score

  return average_scores

```

## Number of Kids that increased before and after

For people that took more than one survey, how many gave an increased rating? Let's look per which two they took and amount gave_increased_rating by question.

```python
# function to count
def count_increased_scores(df1, df2):
  """
  This function counts the number of participants who had a higher question score in df2 compared to df1.

  Args:
      df1 (pd.DataFrame): The first DataFrame containing participant data.
      df2 (pd.DataFrame): The second DataFrame containing participant data.

  Returns:
      pd.DataFrame: A DataFrame with two columns: "Question Number" and "Increased Count".
  """

  increased_counts = {}
  for question_id in df1['question_id'].unique():
    df1_question = df1[df1['question_id'] == question_id]
    df2_question = df2[df2['question_id'] == question_id]

    # Merge the dataframes based on participant code
    merged_df = pd.merge(df1_question, df2_question, on='PARTICIPANT CODE')

    # Count the number of participants with increased scores
    # y (df2) is counted if it's greater than x (df1)
    increased_count = merged_df[merged_df['question_score_y'] > merged_df['question_score_x']].shape[0]

    increased_counts[question_id] = increased_count

  # Create a DataFrame with the results
  data = {
      "Question Number": list(increased_counts.keys()),
      "Increased Count": list(increased_counts.values()),
  }
  return pd.DataFrame(data)


```

Count number of participants that claimed each identity. Participants are allowed to claim more than one.

```python
ethnicity_list = ["white", "latin", "black", "asian", "islander", "unknown"]


def count_ethnicities(df_demos, ethnicity_list):
  # Initialize a dictionary to store counts for each ethnicity
  ethnicity_counts = {eth: 0 for eth in ethnicity_list}

  # Iterate through each row in the DataFrame
  for index, row in df_demos.iterrows():

    # Cast ethnicity column to string (assuming it's not already)
    df_demos['ethnicity'] = df_demos['ethnicity'].astype(str)
    ethnicity = row['ethnicity']

    # Check for each ethnicity in the list (case-insensitive)
    for eth in ethnicity_list:
      if eth.lower() in ethnicity.lower(): # if ethnicity contained anywhere within the string, will count it
        ethnicity_counts[eth] += 1

  # Create a DataFrame from the dictionary
  df_counts = pd.DataFrame.from_dict(ethnicity_counts, orient='index', columns=['Count'])

  return df_counts

ethnicity_counts = count_ethnicities(df_demos.copy(), ethnicity_list.copy())
print(ethnicity_counts)
```
